```python
#panda is a library and it contain the function reloaded to the database
import pandas as pd
```

```python
#take the dataset varible
#read_csv function was call by the pd class
#pass the parameter as data set fill
dataset=pd.read_csv("Salary_Data.csv")
```

```python
#all the data are assigned by dataset
dataset
```

```python
#input output split
```

```python
#independent=input
#which column you want; variable[["coloumn name"]]
independent=dataset[["YearsExperience"]]
independent
```

```python
#dependent=output
#which coloumn you want; variable[["coloumn"]]
dependent=dataset[["Salary"]]
dependent
```

```python
#split train and test
```

```python
#sklearn provide machine learning algorithm
#from folter.inside folter import class
#pass the parameter ;  output is four list of answer, so i have to assign four input variable

from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(independent, dependent, test_size=0.30, random_state=0)
```

```python
#x_train,x_test=input
#y_train,y_test=output
```

```python
# Model creation
```

```python
#from folder.folter inside import class
# linearRegression has some functions are assigned on regressor
# regressor kulla irukira fit function call pannen(pass the model)

from sklearn.linear_model import LinearRegression
regressor=LinearRegression()
regressor.fit(X_train,y_train)                 #this holds weight,bias and formula
```

```python
# coef_ word is used to calulate weight

weight=regressor.coef_
weight
```

```python
# intercept_word is used to calculate bias
bais=regressor.intercept_
bais
```

```python
#test set
```

```python
# cross check of test set
y_pred=regressor.predict(X_test)
```

```python
#Evalution matrics
```

```python
#acutual output and pridicted output assign r_score
from sklearn.metrics import r2_score
r_score=r2_score(y_test,y_pred)
```

```python
r_score             #nearly 1 good model
                    #nearly 0 poor model
```

```python
#if your model is good then save the model
```

```python
import pickle                                    #used to save model
filename="finalized_model_linear.sav"            #dammy file name,   pickel extention is .sav
```

```python
#model saved
pickle.dump(regressor,open(filename,'wb'))       #wb-write bit
```

```python
#load the save model
#the saved fill is open then give input and check it
loaded_model=pickle.load(open("finalized_model_linear.sav",'rb'))
result=loaded_model.predict([[15]])              # individual value get panna use two close bracket
```
